```python
import os
import pandas as pd

from utils.read_input import read_input

data = [

]

datasets = ['a', 'b', 'x']

for dataset in datasets:

    folder = f"../datasets/{dataset}"

    for filename in sorted(os.listdir(folder)):
        input = read_input(f"{folder}/{filename}")

        avg_items_per_order = sum(len(o) for o in input.orders) / len(input.orders)

        wave_size_by_avg_items_per_order = (input.ub - input.lb) / avg_items_per_order

        data.append({
            'dataset': dataset,
            'filename': filename,
            'nOrders': input.nOrders,
            'nItems': input.nItems,
            'nAisles': input.nAisles,
            'avg_items_per_order': avg_items_per_order,
            'lb': input.lb,
            'ub': input.ub,
            'wave_size_by_avg_items_per_order': wave_size_by_avg_items_per_order
        })

df = pd.DataFrame(data)

df = df.rename(columns={
    'dataset': 'Dataset',
    'filename': 'Instance',
    'nOrders': 'Orders',
    'nItems': 'Items',
    'nAisles': 'Aisles',
    'avg_items_per_order': 'Avg Items/Order',
    'lb': 'Lower Bound',
    'ub': 'Upper Bound',
    'wave_size_by_avg_items_per_order': 'Wave Size by Avg Items/Order'
})

df
```

```python
for dataset in datasets:

    folder = f"../datasets/{dataset}"

    for filename in sorted(os.listdir(folder)):
        input = read_input(f"{folder}/{filename}")
        
        demand: dict[int, int] = {}
        
        for order in input.orders:
            
            for idx, qnt in order.items():
                
                if idx not in demand:
                    demand[idx] = 0
                
                demand[idx] += qnt

        stock: dict[int, int] = {}
        
        for aisle in input.aisles:
            
            for idx, qnt in aisle.items():
                
                if idx not in stock:
                    stock[idx] = 0
                
                stock[idx] += qnt
                
        count = 0
                
        for idx in range(input.nItems):
            if idx not in demand:
                demand[idx] = 0
            if idx not in stock:
                stock[idx] = 0
            
            if demand[idx] > stock[idx]:
                count += 1

        print(f"{dataset} - {filename} - {count} items with demand greater than stock")
    
```
