```python
import os
import pandas as pd

from utils.read_input import read_input

data = [

]

datasets = ['a', 'b', 'x']

for dataset in datasets:

    folder = f"../datasets/{dataset}"

    for filename in sorted(os.listdir(folder)):
        nOrders, nItems, nAisles, orders, aisles, lb, ub = read_input(f"{folder}/{filename}")

        avg_items_per_order = sum(len(o) for o in orders) / len(orders)

        wave_size_by_avg_items_per_order = (ub - lb) / avg_items_per_order

        data.append({
            'dataset': dataset,
            'filename': filename,
            'nOrders': nOrders,
            'nItems': nItems,
            'nAisles': nAisles,
            'avg_items_per_order': avg_items_per_order,
            'lb': lb,
            'ub': ub,
            'wave_size_by_avg_items_per_order': wave_size_by_avg_items_per_order
        })

df = pd.DataFrame(data)

df = df.rename(columns={
    'dataset': 'Dataset',
    'filename': 'Instance',
    'nOrders': 'Orders',
    'nItems': 'Items',
    'nAisles': 'Aisles',
    'avg_items_per_order': 'Avg Items/Order',
    'lb': 'Lower Bound',
    'ub': 'Upper Bound',
    'wave_size_by_avg_items_per_order': 'Wave Size by Avg Items/Order'
})

df
```

```python
summary = df.groupby('Dataset').agg(
    Avg_Orders=('Orders', 'mean'),
    Avg_Items=('Items', 'mean'),
    Avg_Items_Per_Order=('Avg Items/Order', 'mean'),
    Avg_Lower_Bound=('Lower Bound', 'mean'),
    Avg_Upper_Bound=('Upper Bound', 'mean'),
).round(2).reset_index()

summary.columns = ['Dataset', 'Avg Orders', 'Avg Items', 'Avg Items/Order', 'Avg Lower Bound', 'Avg Upper Bound']

summary
```

```python

aisle_data = []

for dataset in datasets:
    folder = f"../datasets/{dataset}"

    for filename in sorted(os.listdir(folder)):
        nOrders, nItems, nAisles, orders, aisles, lb, ub = read_input(f"{folder}/{filename}")

        items_per_aisle = [len(aisle) for aisle in aisles]

        avg_items = sum(items_per_aisle) / len(items_per_aisle)
        avg_sparsity = avg_items / nItems

        aisle_data.append({
            'Dataset': dataset,
            'Instance': filename,
            'Total Items': nItems,
            'Aisles': nAisles,
            'Avg Items/Aisle': round(avg_items, 2),
            'Min Items/Aisle': min(items_per_aisle),
            'Max Items/Aisle': max(items_per_aisle),
            'Avg Sparsity': round(avg_sparsity, 4),
        })

df_aisles = pd.DataFrame(aisle_data)
df_aisles
```

```python

df_aisles.groupby('Dataset').agg(
    Avg_Items_Per_Aisle=('Avg Items/Aisle', 'mean'),
    Min_Items_Per_Aisle=('Min Items/Aisle', 'min'),
    Max_Items_Per_Aisle=('Max Items/Aisle', 'max'),
    Avg_Sparsity=('Avg Sparsity', 'mean'),
    Min_Sparsity=('Avg Sparsity', 'min'),
    Max_Sparsity=('Avg Sparsity', 'max'),
).round(4).reset_index().rename(columns={
    'Avg_Items_Per_Aisle': 'Avg Items/Aisle',
    'Min_Items_Per_Aisle': 'Min Items/Aisle',
    'Max_Items_Per_Aisle': 'Max Items/Aisle',
    'Avg_Sparsity': 'Avg Sparsity',
    'Min_Sparsity': 'Min Sparsity',
    'Max_Sparsity': 'Max Sparsity',
})
```
